```python
from bs4 import BeautifulSoup
import requests
import csv
import re
#inspirado en el video de  https://www.youtube.com/watch?v=RjfqdJEwWyU&t=424s
# extrae precio (en cop) y calificuación del usuario (1-5) de un artículo de página web de linio. 
def price_urate(url):
    resultado = requests.get(url)
    contenido =resultado.text
    soup = BeautifulSoup(contenido, "lxml")
    try:
        price=soup.find("span", class_="price-main-md").get_text()
        j=soup.find("span", class_="rating-average body-base-sm").get_text()
    except:
        price=soup.find("span", class_="price-main-md").get_text()
        j=0       
    return re.sub("\n","",price)[1:],j
#Enlace usados para iterar
#https://www.linio.com.co/search?scroll=&q=camisas%20hombre&page=

f = csv.writer(open('precio_urate.csv', 'w'))
f.writerow(['urate', 'price'])
for i in range(17):
    url_main="https://www.linio.com.co/search?scroll=&q=camisetas&sortBy=price_asc&page="+str(i+1)
    resultado_main = requests.get(url_main)
    contenido_main =resultado_main.text
    soup_main = BeautifulSoup(contenido_main, "lxml")
    url_products=soup_main.find_all("a",class_="rating-container col-12 pl-0")
    r=0
    for url in url_products:
        
        price,urate=price_urate("https://www.linio.com.co"+url.get("href"))
        f.writerow([urate,price])
        print(urate, price)
```
